```python
data['NOTA_FINALA_EA'] = data.apply(lambda row: row['NOTA_CONTESTATIE_EA'] if row['CONTESTATIE_EA'].lower() == 'da' else row['NOTA_EA'], axis=1)
data['NOTA_FINALA_EB'] = data.apply(lambda row: row['NOTA_CONTESTATIE_EB'] if row['CONTESTATIE_EB'].lower() == 'da' else row['NOTA_EB'], axis=1)
data['NOTA_FINALA_EC'] = data.apply(lambda row: row['NOTA_CONTESTATIE_EC'] if row['CONTESTATIE_EC'].lower() == 'da' else row['NOTA_EC'], axis=1)
data['NOTA_FINALA_ED'] = data.apply(lambda row: row['NOTA_CONTESTATIE_ED'] if row['CONTESTATIE_ED'].lower() == 'da' else row['NOTA_ED'], axis=1)
data.drop(columns=['NOTA_EA', 'CONTESTATIE_EA', 'NOTA_CONTESTATIE_EA','NOTA_EB', 'CONTESTATIE_EB', 'NOTA_CONTESTATIE_EB','NOTA_EC', 'CONTESTATIE_EC', 'NOTA_CONTESTATIE_EC','NOTA_ED', 'CONTESTATIE_ED', 'NOTA_CONTESTATIE_ED'], inplace=True)
data.rename(columns={'Fileira': 'Filiera'}, inplace=True)
data.head()

# Save the DataFrame as a CSV file
csv_file = "./BAC/bac2019.csv"
data.to_csv(csv_file, index=False)  # index=False prevents pandas from writing row numbers

print(f"Data has been saved to {csv_file}")
```

```python
print("Extracting and processing tables from PDF...")

# Extract tables from all pages of the PDF
df_list = read_pdf("./Admiteri/UPB_ACS_2014.pdf", pages='all', multiple_tables=True)

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Replace NaN values with empty strings
combined_df = combined_df.fillna('')

# Convert rows to strings and clean up multiple commas
combined_df = combined_df.astype(str).apply(lambda x: ','.join(x).replace(',,', ',').strip(','), axis=1)

# Adjust rows after the 32nd entry
def adjust_entry(entry):
    values = entry.split(',')
    if len(values) > 6:  # Ensure there are at least 7 columns
        position = values[6]  # Extract the 7th column value (index 6)
        remaining = values[:6] + values[7:]  # Exclude the positional value from its original place
        adjusted_entry = [position] + remaining
        return ','.join([v for v in adjusted_entry if v])  # Remove empty fields (extra commas)
    return entry

# Remove ".0" decimals from entries
def remove_decimal(entry):
    values = entry.split(',')
    values[0] = str(int(float(values[0]))) if values[0].replace('.', '', 1).isdigit() and values[0].endswith('.0') else values[0]
    return ','.join(values)

# Apply adjustments only to rows after the 32nd entry
combined_df.iloc[33:] = combined_df.iloc[33:].apply(adjust_entry).apply(remove_decimal)

# Add proper headers
headers = [
    "Nr.", "Nume", "I.T.", "Prenume", "Optiuni", 
    "BAC", "Proba 1", "Proba 2", "Medie Probe", "Medie GEN"
]

# Split the cleaned rows back into columns
combined_df = combined_df.apply(lambda x: pd.Series(x.split(',')))

# Trim the number of columns to match the headers if there are extra columns
combined_df = combined_df.iloc[:, :len(headers)]  # Keep only the first 10 columns
combined_df.columns = headers  # Assign the headers
combined_df = combined_df.drop(index=0)

# Save the cleaned DataFrame to a CSV file without quotes
output_file = "./Admiteri CSV/UPB_ACS_raw_2014.csv"
combined_df.to_csv(output_file, index=False, quoting=csv.QUOTE_NONE, escapechar=' ')

print(f"Cleaned and processed data saved to {output_file}.")
```

```python
print("Extracting GEOGRAFIE results to CSV...")

# Extract tables from all pages of the PDF
pdf_path = "./Admiteri/GEOGRAFIE_2024.pdf"
df_list = read_pdf(pdf_path, pages='all', multiple_tables=True)

# Combine all extracted tables into a single DataFrame
df = pd.concat(df_list, ignore_index=True)

# Define the headers
headers = ["Nr.", "Dosar", "Medie BAC"]

# Drop completely empty rows and reset index
df = df.dropna(how="all").reset_index(drop=True)

# Extract relevant columns
# Adjust the slicing based on the actual structure of the data
df_cleaned = df.iloc[:, [1, 3, 7]]  # Assuming relevant columns are at indices 1, 3, and 7

# Assign headers
df_cleaned.columns = headers

# Clean up unnecessary spaces and remove completely empty rows
df_cleaned = df_cleaned.applymap(lambda x: str(x).strip() if pd.notnull(x) else "").replace("", None)
df_cleaned = df_cleaned.dropna(how="all")  # Drop rows that are completely empty
df_cleaned = df_cleaned.drop(index=0)

# Save the cleaned DataFrame to a new CSV file
output_file = "./Admiteri CSV/GEOGRAFIE_raw_2024.csv"
df_cleaned.to_csv(output_file, index=False)

print(f"Cleaned and processed data saved to {output_file}.")
```
